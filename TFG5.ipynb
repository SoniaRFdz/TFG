```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Cargar datos
df = pd.read_csv('datos_flujo_laminar_simulado.csv')

# Configuración de estilo para gráficos
sns.set(style="whitegrid")

#Gráfico de Caudal vs. Diámetro para Distintas Presiones
plt.figure(figsize=(10, 6))
for presion in df['Presion (Pa)'].unique():
    subset = df[df['Presion (Pa)'] == presion]
    plt.plot(subset['Diametro (m)'], subset['Flujo Volumetrico (m³/s)'], label=f'Presión: {presion:.0f} Pa')

plt.xlabel('Diámetro (m)')
plt.ylabel('Flujo Volumétrico (m³/s)')
plt.title('Caudal vs. Diámetro para diferentes presiones')
plt.legend(title="Presión")
plt.show()

#Gráfico de Caudal vs. Presión para Distintos Diámetros
plt.figure(figsize=(10, 6))
for diametro in df['Diametro (m)'].unique():
    subset = df[df['Diametro (m)'] == diametro]
    plt.plot(subset['Presion (Pa)'], subset['Flujo Volumetrico (m³/s)'], label=f'Diámetro: {diametro:.2f} m')

plt.xlabel('Presión (Pa)')
plt.ylabel('Flujo Volumétrico (m³/s)')
plt.title('Caudal vs. Presión para diferentes diámetros')
plt.legend(title="Diámetro")
plt.show()

#Gráfico de Número de Reynolds vs. Diámetro
plt.figure(figsize=(10, 6))
for velocidad in df['Velocidad (m/s)'].unique():
    subset = df[df['Velocidad (m/s)'] == velocidad]
    plt.plot(subset['Diametro (m)'], subset['Reynolds'], label=f'Velocidad: {velocidad:.2f} m/s')

plt.axhline(2000, color='red', linestyle='--', label='Límite de flujo laminar (Re = 2000)')
plt.xlabel('Diámetro (m)')
plt.ylabel('Número de Reynolds')
plt.title('Número de Reynolds vs. Diámetro para diferentes velocidades')
plt.legend(title="Velocidad")
plt.show()

```
